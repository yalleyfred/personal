##  Model

```python
X = data.drop('status_type', axis = 1)
y = data['status_type']
```

```python
columns = X.columns
```

```python
scaler = MinMaxScaler()
```

```python
X.head()
```

```python
X = scaler.fit_transform(X)
```

```python
# changing array to Pandas Dataframe
X = pd.DataFrame(X, columns = [columns])
X.head()
```

```python
model = KMeans(n_clusters = 2)
```

```python
model.fit(X)
```

```python
model.inertia_
```

```python
labels = model.labels_
```

```python
labels
```

```python
correct_labels = sum(labels == y)
```

```python
correct_labels/y.size
```

```python
cluster_nums = list(range(2,11,1))
```

```python
for num in cluster_nums:
    model = KMeans(n_clusters = num)
    model.fit(X)
    print('model_inertia: ', model.inertia_)
    labels = model.labels_
    correct_labels = sum(labels == y)
    print('correct_labels/y.size for n = {}: '.format(num), correct_labels/y.size)
```
